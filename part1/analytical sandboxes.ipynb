```python
result_df = pd.merge(df_rents, df_area, left_on='AreaID', right_on='id', how='inner')
```

```python
result_df = pd.merge(result_df, df_trim, left_on='TrimesterID', right_on='id', how='inner')
```

```python
rents_sandbox = result_df[['avg_rent', 'district', 'neighbourhood', 'trimester', 'year']]
```

```python
rents_sandbox = rents_sandbox.groupby(['year','district','neighbourhood'])['avg_rent'].mean().reset_index()
rents_sandbox
```

```python
result_df = pd.merge(df_income, df_area, left_on='AreaID', right_on='id', how='inner')
result_df = pd.merge(result_df, df_year, left_on='YearID', right_on='id', how='inner')
result_df.head()
```

```python
income_sandbox = result_df[['Income','district','neighbourhood', 'year']]
income_sandbox.shape
income_sandbox.head()
```

```python
df_analysis = pd.merge(income_sandbox, rents_sandbox, on=['year', 'district', 'neighbourhood'], how='left')
print(df_analysis.shape)
df_analysis.head()
```

```python
df_analysis[df_analysis.isnull().any(axis=1)].shape
```

```python
# Group by 'neighbourhood' and 'year', calculate the mean of 'avg_rent' for each group
neighborhood_avg_rent = df_analysis.groupby(['neighbourhood', 'year'])['avg_rent'].mean().reset_index()

# Fill NaN values in 'avg_rent' based on the average rent of the same neighborhood in different years
df_analysis['avg_rent'] = df_analysis.apply(
    lambda row: neighborhood_avg_rent.loc[
        (neighborhood_avg_rent['neighbourhood'] == row['neighbourhood']) & 
        (neighborhood_avg_rent['year'] != row['year'])
    ]['avg_rent'].mean() if pd.isna(row['avg_rent']) else row['avg_rent'],
    axis=1
)
```

```python
df_analysis= df_analysis.dropna(subset=['avg_rent'])
df_analysis.shape
```

```python
df_analysis['Income_Rent_Ratio'] = df_analysis['Income_Rent_Ratio'] = df_analysis['Income'] / df_analysis['avg_rent']
df_analysis['District_Avg_Rent'] = df_analysis.groupby(['district', 'year'])['avg_rent'].transform('mean')
df_analysis['District_Income_Variability'] = df_analysis.groupby('district')['Income'].transform('std')
df_analysis.head()
```

```python
df_analysis['Income_Group'] = pd.cut(df_analysis['Income'], bins=[0, 10000, 20000, 30000, 40000, float('inf')], labels=['Low', 'Medium-Low', 'Medium', 'Medium-High', 'High'])
df_analysis.head()
```
